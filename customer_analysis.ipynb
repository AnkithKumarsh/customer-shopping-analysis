```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Review_Rating'] = df['Review_Rating'].fillna(df['Review_Rating'].mean())
```

```python
df['Review_Rating'].isnull().sum()
```

```python
total_revenue = df['Purchase_Amount'].sum()
print("Total Revenue:", total_revenue)
```

```python
category_sales = df.groupby('Category')['Purchase_Amount'].sum()

category_sales.plot(kind='bar', title="Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Purchase_Amount'].mean()

gender_sales.plot(kind='bar', title="Average Spending by Gender")
plt.show()
```

```python
age_sales = df.groupby('Age_Group')['Purchase_Amount'].sum()

age_sales.plot(kind='bar', title="Sales by Age Group")
plt.show()
```

```python
payment = df['Payment_Method'].value_counts()

payment.plot(kind='pie', autopct='%1.1f%%', title="Payment Methods")
plt.show()
```
